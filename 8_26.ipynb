```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python
# Dataset
X = np.array([1, 2, 3, 4, 5]).reshape(-1 , 1)
# reshape(-1 , 1) because we only need one column and put row = -1 so it will adjust automatically but column will be 1 that is called flatten

Y = np.array([35, 37, 45, 50, 60])

# Train Model
model = LinearRegression()
model.fit(X, Y)

# Predictions
y_pred = model.predict(X)
print(y_pred)

# Plot Result
plt.scatter(X, Y , color = 'blue' , label = 'Actual Data')
plt.plot(X, y_pred, color = 'red' , label = 'Predicted Data')
plt.xlabel('Hours Studied')
plt.ylabel('Marks')
plt.show()

print('Intercept B0: ', model.intercept_)
print('Coefficient B1: ', model.coef_)

# Comments Blue points are scatter plot, our model calculated best fit line acc to it.
# on an avg 6.3 marks is increasing after increasing 1 unit of x that is slope
# intercept: y = 0 still student got 26.5 marks
```

```python
# Multiple Linear Regression
# Predict Y using Multiple pridictors(x)
# Y = B0 + B1x1 + B2x2 + B3x3 + ... + Bnxn + E

from sklearn.linear_model import LinearRegression
```

```python
# Example Dataset
data = {
    'size_sqft': [1000, 1500, 2000, 2500, 3000],
    'bedrooms': [2, 3, 3, 4, 4],
    'location_score': [5 , 6 , 7 , 8 , 9],
    'price': [200000, 250000, 300000, 350000, 400000]
}

df = pd.DataFrame(data)
```

```python
Xml = df[['size_sqft', 'bedrooms', 'location_score']]
Yml = df['price']

# Train Model
modelML = LinearRegression()
modelML.fit(Xml, Yml)

#Prediction
y_predML = modelML.predict(Xml)
print(y_predML)
print('Intercept B0: ', modelML.intercept_)
print('Coefficient B1 , B2 , B3: ', modelML.coef_)
```

```python
# Prediction Result
plt.scatter(Yml, y_predML , color = 'blue' , label = 'Predicted Data VS Actual Data')
plt.plot([Yml.min(), Yml.max()] , [Yml.min() , Y.max()], color = 'red' , linestyle = '--') # [Yml.min(), Yml.max()] why?? =>
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.legend()
plt.show()
# Data is Overfitted thats why actual == predict
```
