```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/sharmaroshan/Heart-UCI-Dataset/master/heart.csv')
```

```python
df.columns = ['Age' , 'Sex' , 'ChestPainType' , 'RestBP' , 'Cholesterol' , 'FBS' , 'RestECG' , 'MaxHR' , 'ExerciseAngina' , 'Oldpeak' , 'ST_Slope' , 'NumVesselsFluoro' , 'Thal' , 'Target']
# OR df.rename(columns = '')
df.head()
```

```python
# One Hot Encoding
df_encoded = pd.get_dummies(df , columns = ['ChestPainType'] , drop_first= True)
df_encoded.head()

# if hum ml model ko 1 2 3 numeric ki form mein data deingy tou wo lessthan or greater than ki base pr selection karega jou ke hamein nahi karwani isliye we are doing one hot encoding isme just ye hota hai ke types ko wo separate kardeta means agar patient ko koi pain type hai tou true otherwise false and isme one more thing ke jitne bhi columns honge ussay ek kam ki dummy banate hain usually we drop the first one by drop_first = True
```

```python
# HeatMap of NUmeric Data Only
# Correlation Matrix
numeric_df = df.select_dtypes(include = [np.number])
plt.title("HeatMap of Numerical Data in HeartUCI")
plt.figure(figsize=(14 , 10))
sns.heatmap(numeric_df.corr() , annot = True , cmap = 'Blues')
plt.show()
```

```python
encoded2 = pd.get_dummies(df , columns = [ 'ChestPainType' , 'Thal'] , drop_first= True)
encoded2.head()
```

```python
# Correlation Matrix of encoded2

plt.figure(figsize=(14 , 10))
plt.title('HeatMap of DF_Encoded2')
sns.heatmap(encoded2.corr() , annot = True , cmap = 'Oranges')
plt.show()
```
