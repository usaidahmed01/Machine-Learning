```python
from sklearn.datasets import make_blobs, make_circles
import pandas as pd
from sklearn.svm import SVC
import matplotlib.pyplot as plt
```

```python
X,y=make_blobs(n_samples=50,centers=2,random_state=6)
model=SVC(kernel='linear')
model.fit(X,y)
```

```python
X[:,0]
```

```python
plt.scatter(X[: , 0], X[: , 1], c= y, cmap= 'coolwarm')
w = model.coef_[0]
b = model.intercept_[0] # bias and y-intercept
print(w)
print(b)

x_values = [min(X[:,0]), max(X[:,0])]
y_values = [(-w[0] * x - b) / w[1] for x in x_values] # Formula of SVM Line
plt.plot(x_values, y_values, 'k-',label='Decision Boundary')

y_values_upper = [(-w[0] * x - b + 1) / w[1] for x in x_values]
y_values_down = [(-w[0] * x - b - 1) / w[1] for x in x_values]
plt.plot(x_values, y_values_upper, 'k--',label='Upper Margin +1')
plt.plot(x_values, y_values_down, 'k--',label='Lower Margin -1')

plt.scatter(model.support_vectors_[: , 0], model.support_vectors_[: , 1], s= 120, facecolors='none', edgecolors='k')# Support Vectors


plt.legend()
plt.show()
```

```python
'''
After training the SVM gives you the equation of the line

w1x1 + w2x2 + b = 0
in our case
w1x + w2y + b = 0
y = (-w1x - b) / w2

Decision Boundary w1x + w2y + b = 0 => y = (-w1x - b) / w2
Margin Lines
Upper Margin w1x + w2y + b = 1 => y = (-w1x - b + 1) / w2
Lower Margin w1x + w2y + b = -1 => y = (-w1x - b - 1) / w2
'''
```

```python
X , y = make_circles(n_samples = 200 , random_state=  0 , noise = 0.1)
plt.scatter(X[:, 0], X[: , 1], c = y  , cmap = 'coolwarm')
plt.title("Non-Linear Data - Circular Pattern")
plt.show()
```

```python
modelRBF = SVC(kernel = 'rbf' , C = 1)
modelRBF.fit(X, y)
# RBF=> Radial Basis Function
# Automatically maps data into a higher dimensions where it becomes linear
# like agar circular data hou or usko 2 groups mein separate karna hou tou rbf just increase the dimension but separate in a nice way

# Theres also the kernel called Polynomial that just draw the curve

# C = 1 normal , C = 0 freehand , C = 2 tightly
```

```python
from sklearn.metrics import accuracy_score
yr_pred = model.predict(X)
print("RBF Kernel Accuracy ", accuracy_score(y , yr_pred))
```

```python
plt.scatter(X[: , 0] , X[: , 1] , c = yr_pred , cmap = 'coolwarm')
plt.title("SVM with RBF Kernel")
plt.show()
```

```python
from sklearn.datasets import make_moons
Xr , yr = make_moons(n_samples=200 , noise = 0.1 , random_state= 0)
model = SVC(kernel = 'rbf' , gamma = 0.5)
model.fit(Xr , yr)
yrr_pred = model.predict(Xr)
print("RBF Kernel Accuracy ", accuracy_score(yr , yrr_pred))
```

```python
plt.scatter(Xr[: , 0] , Xr[: , 1] , c = yrr_pred , cmap = 'coolwarm')
plt.title("SVM with RBF Kernel")
plt.show()
```

```python
Xp , yp = make_circles(n_samples = 200 , factor= 0.5 , noise = 0.05)
model = SVC(kernel = 'poly' , degree = 3 , C = 1)
model.fit(Xp , yp)
yp_pred = model.predict(Xp)
print("Accuracy of Poly Kernel " , accuracy_score(yp , yp_pred))
```

```python
plt.scatter(Xp[: , 0] , Xp[: , 1] , c = yp_pred , cmap = 'coolwarm')
plt.title("Poly Kernel")
plt.show()
```
