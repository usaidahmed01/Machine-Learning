```python
from sklearn.datasets import make_blobs
import pandas as pd
from sklearn.svm import SVC
import matplotlib.pyplot as plt
```

```python
X,y=make_blobs(n_samples=50,centers=2,random_state=6)
model=SVC(kernel='linear')
model.fit(X,y)
```

```python
X[:,0]
```

```python
plt.scatter(X[: , 0], X[: , 1], c= y, cmap= 'coolwarm')
w = model.coef_[0]
b = model.intercept_[0] # bias and y-intercept
print(w)
print(b)

x_values = [min(X[:,0]), max(X[:,0])]
y_values = [(-w[0] * x - b) / w[1] for x in x_values] # Formula of SVM Line
plt.plot(x_values, y_values, 'k-',label='Decision Boundary')

y_values_upper = [(-w[0] * x - b + 1) / w[1] for x in x_values]
y_values_down = [(-w[0] * x - b - 1) / w[1] for x in x_values]
plt.plot(x_values, y_values_upper, 'k--',label='Upper Margin +1')
plt.plot(x_values, y_values_down, 'k--',label='Lower Margin -1')

plt.scatter(model.support_vectors_[: , 0], model.support_vectors_[: , 1], s= 120, facecolors='none', edgecolors='k')# Support Vectors


plt.legend()
plt.show()
```

```python
'''
After training the SVM gives you the equation of the line

w1x1 + w2x2 + b = 0
in our case
w1x + w2y + b = 0
y = (-w1x - b) / w2

Decision Boundary w1x + w2y + b = 0 => y = (-w1x - b) / w2
Margin Lines
Upper Margin w1x + w2y + b = 1 => y = (-w1x - b + 1) / w2
Lower Margin w1x + w2y + b = -1 => y = (-w1x - b - 1) / w2
'''
```
